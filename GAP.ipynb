```python
import cvxpy as cp
import numpy as np
```

```python
c = np.array([2, 11, 10])
mi = np.array([1, 2, 3])
vi = np.array([1, 4, 2])
M = np.array([1, 5, 8])
V = np.array([1, 9, 6])
```

```python
#n = len(M) #len(V)
#m = len(mi) #len(v)
n = 10
m = 50
mi = np.random.rand(m)
vi = np.random.rand(m)
M = np.random.rand(n)*0.2*m
V = np.random.rand(n)*0.2*m
c = np.random.rand(n)
x = cp.Variable((m, n), boolean=True)
y = cp.Variable(n, boolean=True)

cons = [mi@x[:,i] <= M[i]*y[i] for i in range(n)] + \
        [vi@x[:,i] <= V[i]*y[i] for i in range(n)] + \
        [sum(x[i,:]) == 1 for i in range(m)]

prob = cp.Problem(cp.Minimize(c.T@y), cons)
```
